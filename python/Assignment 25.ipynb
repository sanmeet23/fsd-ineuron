Q2 - What is the relationship between generators and iterators?

An iterator is an object that can be iterated over, meaning that you can use a for loop to access its elements one at a time. Generators are created using the yield keyword. When a generator function is called, it returns an iterator object. The iterator object can then be used in a for loop to iterate over the sequence of values generated by the generator function.

Q4 - What is the purpose of a yield statement?

The yield statement of a function returns a generator object rather than just returning a value to the call of the function that contains the statement. The Python interpreter halts execution of the function when a yield statement is encountered when calling a function in a programme.
